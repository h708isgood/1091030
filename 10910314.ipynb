```
import tensorflow as tf
print(tf.__version__)
x=tf.random.normal([2,784])
w1=tf.Variable(tf.random.truncated_normal([784, 256], stddev=0.1))
b1=tf.Variable( tf.zeros([256]) ) #b1 
o1=tf.matmul(x,w1)+b1 # x@W+b  x:784, W:784x256 ==> 256, 
o1=tf.nn.relu(o1) #sigma, relu= activive function 

```

```
x=tf.random.normal( [4, 28*28] )
from tensorflow.keras import layers

#等於呼叫Dense的建構子__init() 得到物件 fc 
#此層有512個類神經元
f1=layers.Dense(512, activation=tf.nn.relu)
f2=layers.Dense(256, activation=tf.nn.relu)
f3=layers.Dense(128, activation=tf.nn.relu)
f4=layers.Dense(64, activation=tf.nn.relu)
f5=layers.Dense(32, activation=tf.nn.relu)
f6=layers.Dense(16, activation=tf.nn.relu)
f7=layers.Dense(10, activation=tf.nn.softmax)

h1=f1(x)  #透過__call__,將fc當function使用, 呼叫__call__(x) 
h2=f2(h1) 
h3=f3(h2)
h4=f4(h3)
h5=f5(h4)
h6=f6(h5)
h7=f7(h6)
f7.kernel #16x10 
f6.kernel.shape #32x16
f6.bias.shape  #[16]
f1.kernel.shape  #TensorShape([784, 512])
f2.kernel.shape  #TensorShape([512, 256])
print(f7.bias, f7.kernel)
```
